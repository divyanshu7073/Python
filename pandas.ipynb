
# Some Preprocessing techinques

```python
a = pd.Series([1,np.nan,2,3,4,5,np.nan,7,8,np.nan])
print(a)
```

```python
a.isnull()
```

```python
dict1 = {
    0:[1,4,7,np.nan],
    1:[2,5,np.nan,np.nan],
    2:[3,np.nan,np.nan,np.nan],
    3:[np.nan,np.nan,np.nan,np.nan]
}

df = pd.DataFrame(dict1)
df
```

```python
##drops an entire row if one nan value is present
df.dropna()
```

```python
# deletes the rows in which all the values are null
df.dropna(how='all')
```

```python
# deletes the columns in which all the values are null
df.dropna(how='all',axis = 1)
```

# Filling the null values
* syntax --- > df.fillna(val)
* returns a new DataFrame

```python
df_fill = df.fillna(0)
```

```python
df_fill
```

```python
# to fill particular values in particular columns
df_fill = df.fillna({1:5,3:7})
df_fill
```

# df.mean()
* gives the mean
* to fill the missing values by the mean values we use mean()

```python
df_mean = df.fillna(df.mean())
df_mean
```

# .replace(old_val,new_val)
* replaces the old values with new values
* values can be anything.

```python
df.replace(3,5)
```

```python
df.replace(np.nan,5)
```

```python
df.replace(np.nan,df.mean())
```
